Now state the differences between a zip and enumerate iterator.

Enumerate() method adds a counter to an iterable and returns it in a form of enumerate object. 
This enumerate object can then be used directly in 'for loops' or be converted into a list of tuples using list() method.
In simpler words, it allows us to loop over a list and have an automatic counter.
    On the other hand, zip() is used to map the similar index of multiple containers/list so that they can be used just using as single entity.

Count the number of words in the file

```python
num_words = 0
with open('poem.txt') as f:
    for line in f:
        words = line.split()
        num_words += len(words)
    
print("Number of words:")
print(num_words)
```

## Dictionaries associated with lists

Create a list of keys and values

```python
x = {'DSB': 'dsb@iiitg', 'Rakesh': 'rakesh@iiitg'}
x['DSB']
```

```python
def getList(x): 
    list = [] 
    for key in x.keys(): 
        list.append(key) 
          
    return list

x = {'DSB': 'dsb@iiitg', 'Rakesh': 'rakesh@iiitg'}
print(getList(x))
```

```python
def getList(x): 
    list = [] 
    for value in x.values(): 
        list.append(value) 
          
    return list

x = {'DSB': 'dsb@iiitg', 'Rakesh': 'rakesh@iiitg'}
print(getList(x))
```

```python
x['GB'] = None
```

Iterate over all the keys

```python
# Iterating over keys 
for name in x: 
    print(name) 
```

Iterate over all the values

```python
for mail_ID in x.values():
    
    print(mail_ID)
```

Iterate over all the key-vlaue pairs in the list

```python
for name, mail_ID in x.items():
    
    print(name, ":", mail_ID) 
```

```python
class Test:
    def __init__(self):
        self.L1 = [1, 5, 9, 2, 7]
        
        self.L2 = [2, 3, 20, 8, 30]
p2=Test()

result = list(map(min, zip(p2.L1, p2.L2)))
print (result)
```
